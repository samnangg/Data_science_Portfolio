2. Solve the following system of linear equations using Gaussian elimination (Ax=y)¶
x + 2y + 3x = 9

2x - y + 2z = −3

3x + 3y + z = 5

```r
# If you haven’t insall matlib library run this
#install.packages("rgl", repos="http://R-Forge.R-project.org")
#install.packages("matlib")
```

```r
library(matlib)
A <- matrix(c(1, 2, 3,
            2, 1, 2,
            3,  3, 1), 3, 3, byrow=TRUE)
b <- c(9, -3, 5)
ans <- gaussianElimination(A, b)
ans
paste("answer is x =",ans[10],' y = ', ans[11], ' z =',ans[12])
```

```r
# a. Data Source detail(Ex: Link) 
# Dataset on Novel Corona Virus Disease 2019 in India
# Author: sudalairajkumar
# URL : https://www.kaggle.com/sudalairajkumar/covid19-in-india
library(tidyverse)
dataset <- read_csv("../input/covid19-in-india/StatewiseTestingDetails.csv")
covidData <- dataset
covidData
```

```r
# b. Explain the Unit & Necessity of each variable 
# This dataset have 5 variables:
# - Date: Date of observation
# - State: State in India that do the covid tests
# - TotalSamples: Cumulative number of total samples tested till the given date
# - Negative: Cumulative number of negative samples till the given date
# - Positive: Cumulative number of positive samples till the given date
names(covidData)
```

```r
# c. Find the missing values(rows & columns) and replace them with mean(Tidy Dataset) 
covidData[] <- lapply(covidData, function(x) { 
  x[is.na(x)] <- mean(x, na.rm = TRUE)
  x
})
covidData
```

```r
# d. Generate the two new variables(Var1:Mean, Var2: Median from available variable) 
covidData %>% mutate(Mean=mean(Positive),Median=median(Positive))
```

```r
# e. Rename the two existing variables 
covidData %>% rename(N=Negative,P=Positive,)
```

```r
# f. Create a plot using following instructions (using 7 layers of Grammar of Graphics)
```

```r
# i. Choose x and y axis(aes) 
# Choose Date as X and Positive case of covid as Y
data <- dataset %>% group_by(Date) %>% count(Date, wt = Positive)
Date <- data[['Date']]
Positive <- data[['n']]
data
```

```r
# ii. geom_point() - specify the parameters, size : 5, color: red, alpha: ⅕ 
library(ggplot2)
gp = geom_point(alpha=1/5, color="red", size=5)
gp
p <- ggplot(x = data, mapping = aes(x = Date, y = Positive)) + gp
p
```

```r
# iii. Use Facet grid, cartesian coordinates & geom_smooth() 

# facet_grid
p + facet_grid(x)
```

```r
#geom_smooth
p + geom_smooth()
```

```r
# iv. Assign the title to x, y and graph 
p <- p + ggtitle("Number of Positive Covid-19 case in India") +
  xlab("Date") + ylab("Number Os Positive Cases")
p
```
